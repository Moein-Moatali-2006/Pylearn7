```python
import cv2
import numpy as np
import matplotlib.pyplot as plt 
```

```python
image = cv2.imread("input\mandrill_male.jpg")
R,G,B = cv2.split(image)
# R=R.astype(float)
# G=G.astype(float)
# B=B.astype(float)
result = np.zeros(image.shape[:2],dtype=np.uint8)

for i in range(image.shape[0]):
    for j in range(image.shape[1]):
        # result[i,j] = (R[i,j] + B[i,j] + G[i,j]) // 3
        result[i,j] = R[i,j]// 3 + B[i,j]// 3 + G[i,j] // 3

plt.imshow(result,cmap='gray')
plt.show()
```
