```python
import matplotlib.pyplot as plt
from PIL import Image,ImageDraw,ImageFont,ImageOps
```

```python
image = Image.open("original.png")

plt.axis('off')
plt.imshow(image)
plt.show()
```

```python
image_persian = image
draw = ImageDraw.Draw(image_persian)

font = ImageFont.truetype("Samim-Medium.ttf", 40)
text = ""
text = text[::-1]
draw.text((50,50),text,(255,0,0),font)

plt.axis('off')
plt.imshow(image_persian)
plt.show()
```

```python
Hist = image.histogram()

r_hist = Hist[0:256]
g_hist = Hist[256:512]
b_hist = Hist[512:768]



plt.plot(r_hist,color="red")
plt.plot(g_hist,color="green")
plt.plot(b_hist,color="blue")
plt.show()
```

```python
image = image.convert("RGB")
r,g,b = image.split()

r = ImageOps.equalize(r)
g = ImageOps.equalize(g)
b = ImageOps.equalize(b)

equalized_image = Image.merge("RGB",(r,g,b))

plt.axis('off')
plt.imshow(equalized_image)
plt.show()
```

```python
Hist = equalized_image.histogram()

r_hist = Hist[0:256]
g_hist = Hist[256:512]
b_hist = Hist[512:768]



plt.plot(r_hist,color="red")
plt.plot(g_hist,color="green")
plt.plot(b_hist,color="blue")
plt.show()
```

```python
image_gray = image.convert("L")

plt.axis('off')
plt.imshow(image_gray,cmap="gray")
plt.show()
```

```python
Hist = image_gray.histogram()
plt.plot(Hist)
```

```python
image_gray_equalized = ImageOps.equalize(image_gray)

plt.axis('off')
plt.imshow(image_gray_equalized,cmap="gray")
plt.show()
```

```python
Hist = image_gray_equalized.histogram()

plt.plot(Hist)
plt.show()
```
