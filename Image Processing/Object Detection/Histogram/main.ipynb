```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image_original = cv2.imread("original.png",cv2.IMREAD_GRAYSCALE)
hist = cv2.calcHist([image_original], [0], None, [256], [0, 256])

plt.axis('off')
plt.imshow(image_original,cmap='gray')
plt.show()
plt.plot(hist)
```

```python
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
new_image = clahe.apply(image_original)
new_hist = cv2.calcHist([new_image],[0], None, [256], [0, 256])

plt.axis('off')
plt.imshow(new_image,cmap='gray')
plt.show()

plt.plot(new_hist)
```
