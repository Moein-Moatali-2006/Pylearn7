```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image_color = cv2.imread("input\dice_1.png")
image_gray = cv2.cvtColor(image_color,cv2.COLOR_BGR2GRAY)
image_gray[image_gray>230]=0
image_gray=cv2.blur(image_gray,(5,5))

_ , image_threshed = cv2.threshold(image_gray,205,255,cv2.THRESH_BINARY)

image_contours , image_hierarchy = cv2.findContours(image_threshed,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)

parents = dict()
for i,contour in enumerate(image_contours):
    if cv2.contourArea(contour) > 120:
        if image_hierarchy[0][i][3] == -1:
            parents[i] = []
        
        for j in parents:
            if image_hierarchy[0][i][3] == j :
                parents[j].append(i)

for dice in parents:
    if len(parents[dice]) > 0:
        x,y,w,h = cv2.boundingRect(image_contours[dice])
        cv2.rectangle(image_color,[x,y],[x+w,y+h],(0,0,255),1)
        cv2.putText(image_color, f'{len(parents[dice])}', (x + w + 5, y + 20), 
                    cv2.FONT_HERSHEY_COMPLEX, fontScale=0.75, color=(0, 0, 255), thickness=2, lineType=cv2.LINE_AA)
    
        for i in parents[dice]:
            x,y,w,h = cv2.boundingRect(image_contours[i])
            cv2.rectangle(image_color,[x,y],[x+w,y+h],(255,0,0),1)

plt.axis('off')
plt.imshow(cv2.cvtColor(image_color, cv2.COLOR_BGR2RGB))
plt.show()
```
