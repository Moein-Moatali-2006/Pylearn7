```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image_org = cv2.imread("input/dice_2.png")
image = cv2.cvtColor(image_org,cv2.COLOR_BGR2GRAY)
plt.axis('off')
plt.imshow(image,cmap='gray')
plt.show()
```

```python
_ , image_threshed = cv2.threshold(image,127,255,cv2.THRESH_BINARY)
plt.axis('off')
plt.imshow(image_threshed,cmap='gray')
plt.show()
```

```python
contours,_ = cv2.findContours(image_threshed,cv2.RETR_EXTERNAL ,cv2.CHAIN_APPROX_SIMPLE)

result = image_org.copy()

count_dice = 0
for contour in contours:
    if cv2.contourArea(contour) > 1000:
        x,y,w,h = cv2.boundingRect(contour)
        cv2.rectangle(result,(x,y),(x+w,y+h),(0,0,255),3)
        count_dice += 1

cv2.putText(result,f"Count_dice : {str(count_dice)}",(20,30),cv2.FONT_HERSHEY_COMPLEX,1,0,1)
plt.axis('off')
plt.imshow(result)
plt.show()
```
