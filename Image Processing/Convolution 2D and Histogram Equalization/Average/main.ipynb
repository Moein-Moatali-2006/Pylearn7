```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("1.tif")
```

```python
# Avg_kernel 5*5 : 0.04
kernel = np.ones((5,5)) / 0.04
result = cv2.filter2D(image , -1 , kernel)
plt.imshow(result)
plt.show()
```

```python
# Avg_kernel 5*5 : 1.0
kernel = np.ones((5,5)) / 1.0
result = cv2.filter2D(image , -1 , kernel)
plt.imshow(result)
plt.show()
```
